```python
import folium
import pandas as pd
```

```python
input = pd.read_csv('data_train/location_input.csv')
output = pd.read_csv('data_train/location_output.csv')
```

```python
m = folium.Map(location=[21.019,105.819], zoom_start=11.5)
for i, x in input.iterrows():
    folium.Marker([x['latitude'], x['longitude']]).add_to(m)
for i, x in output.iterrows():
    folium.Marker(
        [output.iloc[i]['latitude'], output.iloc[i]['longitude']], 
        icon=folium.Icon(color="red", icon="screenshot", prefix='glyphicon')
    ).add_to(m)
m
```
